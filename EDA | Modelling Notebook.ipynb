## Feature Engineering

- Create a custom transformer that creates new features based on a given feature value

```python
from sklearn.base import BaseEstimator, TransformerMixin
```

```python
class CustomFeatureClassifier(BaseEstimator, TransformerMixin):
    def __init__(self, classify_area=False):
        self.classify_area = classify_area
    
    def fit(self, X, y=None):
        return self
    
    def transform(self, X):
        X["radius_class"] = X["radius_mean"].apply(lambda x: "High" if x > 15 else "Low")
        if self.classify_area:
            X["area_class"] = X["area_mean"].apply(lambda x: "High" if x > 15 else "Low")
        
        return X
```

```python
class CustomDropColumns(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self
    
    def transform(self, X):
        X.drop(columns=["radius_class"])
        return X
```

```python
numeric_features = numeric_cols.copy()

# define Pipeline
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder, FunctionTransformer
from sklearn.compose import ColumnTransformer

numeric_transformer = Pipeline([
    ('scaler', StandardScaler())
])

categorical_transformer = Pipeline([
    ("onehot", OneHotEncoder())
])
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, numeric_features),
        ('new_feature_onehot', OneHotEncoder(), ['radius_class']),  # Step 2: One-hot encoding for new feature
    ]
)

# Function to drop a specific column
def drop_column(X):
    return np.delete(X, [-3], axis=1)

pipeline = Pipeline(steps=[
    ('custom_transform', CustomFeatureClassifier()),  # Step 1: Add new column
    ('preprocessor', preprocessor),
    ('drop_column', FunctionTransformer(drop_column, validate=False))
])
```

```python
transformed_data = pipeline.fit_transform(X_train)
```

## Model

```python
from sklearn.linear_model import LogisticRegression
```

```python
y_train = X_train["diagnosis"]
```

```python
model_lr = LogisticRegression()
model_lr.fit(transformed_data, y_train)
```

## Evaluation

Since the two classes are not significantly different, we can evaluate using the accuracy score

```python
from sklearn.model_selection import cross_val_score
```

```python
scores = cross_val_score(model_lr, transformed_data, y_train, cv=5)
scores.mean()
```
